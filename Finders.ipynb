```python
#Formatação

def parse_galaxy_list(raw_text):
    """
    Parses a plain text list of galaxies into a Python list of (name, ra, dec) tuples.
    """
    galaxies = []
    for line in raw_text.strip().splitlines():
        if not line.strip():
            continue  # Pular as vazias
        parts = line.split()
        if len(parts) < 3:
            print(f"Skipping invalid line: {line}")
            continue
        name = parts[0]
        try:
            ra = float(parts[1])
            dec = float(parts[2])
            galaxies.append((name, ra, dec))
        except ValueError:
            print(f"Skipping invalid numbers in line: {line}")
    return galaxies

# Bota aqui as galaxias

raw_input_text = """
 NGC4658 191.157310 -10.083328
   NGC2775 137.583953   7.037979
   NGC2903 143.042403  21.501831
   NGC4699 192.259168  -8.664333
   NGC4981 197.203149  -6.777515
   NGC4995 197.419343  -7.833410
   NGC5054 199.243655 -16.634561
   NGC5079 199.908528 -12.698959
   NGC4165 183.049182  13.246485
   NGC4193 183.473247  13.172955
   NGC3981 179.031044 -19.895948
   NGC4030 180.098393  -1.100035
   NGC5740 221.101844   1.679810
ESO576-001 197.598995 -21.684189
 PGC053093 223.043393 -10.740294
   NGC5781 224.172071 -17.243358
   NGC3368 161.690350  11.819975
   NGC3521 166.452415  -0.035937
   NGC4799 193.814776   2.896736
   NGC4826 194.181848  21.683075
   NGC3351 160.990618  11.703659
   NGC4450 187.123344  17.084958
   NGC5248 204.383361   8.885187
   NGC3627 170.062528  12.991503
   NGC4380 186.342423  10.016819
   NGC3673 171.303583 -26.736739
   NGC4237 184.297628  15.324096
   NGC5661 217.989218   6.250440
   NGC4568 189.142956  11.238855
   NGC4579 189.431405  11.818196
   NGC4591 189.801880   6.012247
   NGC4501 187.996511  14.420326
   NGC4602 190.153836  -5.132830
    IC0769 183.134680  12.123775

"""

# Aqui é pra ele dar de volta

galaxies = parse_galaxy_list(raw_input_text)

print("\ngalaxies = [")
for name, ra, dec in galaxies:
    print(f'    ("{name}", {ra:.6f}, {dec:.6f}),')
print("]")

# Poderia até fazer pra automatizar tudo mas, assim é melhor pra ficar acordade e revisar o input e output dos negocios
# Tem que copiar e colar no código embaixo
```

```python
#Lista de Galaxias

galaxies = [
    ("NGC4658", 191.157310, -10.083328),
    ("NGC2775", 137.583953, 7.037979),
    ("NGC2903", 143.042403, 21.501831),
    ("NGC4699", 192.259168, -8.664333),
    ("NGC4981", 197.203149, -6.777515),
    ("NGC4995", 197.419343, -7.833410),
    ("NGC5054", 199.243655, -16.634561),
    ("NGC5079", 199.908528, -12.698959),
    ("NGC4165", 183.049182, 13.246485),
    ("NGC4193", 183.473247, 13.172955),
    ("NGC3981", 179.031044, -19.895948),
    ("NGC4030", 180.098393, -1.100035),
    ("NGC5740", 221.101844, 1.679810),
    ("ESO576-001", 197.598995, -21.684189),
    ("PGC053093", 223.043393, -10.740294),
    ("NGC5781", 224.172071, -17.243358),
    ("NGC3368", 161.690350, 11.819975),
    ("NGC3521", 166.452415, -0.035937),
    ("NGC4799", 193.814776, 2.896736),
    ("NGC4826", 194.181848, 21.683075),
    ("NGC3351", 160.990618, 11.703659),
    ("NGC4450", 187.123344, 17.084958),
    ("NGC5248", 204.383361, 8.885187),
    ("NGC3627", 170.062528, 12.991503),
    ("NGC4380", 186.342423, 10.016819),
    ("NGC3673", 171.303583, -26.736739),
    ("NGC4237", 184.297628, 15.324096),
    ("NGC5661", 217.989218, 6.250440),
    ("NGC4568", 189.142956, 11.238855),
    ("NGC4579", 189.431405, 11.818196),
    ("NGC4591", 189.801880, 6.012247),
    ("NGC4501", 187.996511, 14.420326),
    ("NGC4602", 190.153836, -5.132830),
    ("IC0769", 183.134680, 12.123775),
]

output_dir = r"C:\Users\Juan\Downloads\skyview_finders"
os.makedirs(output_dir, exist_ok=True)

surveys_priority = [
    "DSS2 Red",
    "DSS2 Blue",
    "2MASS-K"
]

def save_finder_chart(data, name, ra, dec, output_folder):
    data = np.nan_to_num(data, nan=0.0)
    clipped = np.clip(data, np.percentile(data, 5), np.percentile(data, 99))

    fig, ax = plt.subplots(figsize=(4, 4), dpi=150)
    ax.imshow(clipped, cmap="gray_r", origin="lower")

    ny, nx = clipped.shape
    ax.axhline(ny // 2, color='red', lw=0.5)
    ax.axvline(nx // 2, color='red', lw=0.5)

    ax.set_xticks([])
    ax.set_yticks([])

    title = f"{name}"
    footer = f"RA: {ra:.4f}°, Dec: {dec:.4f}°"
    ax.set_title(title, fontsize=10, fontname="Times New Roman")
    ax.text(0.01, 0.01, footer, transform=ax.transAxes,
            fontsize=7, fontname="Times New Roman", color="black", va='bottom', ha='left')

    output_path = os.path.join(output_folder, f"{name}.jpg")
    plt.savefig(output_path, bbox_inches="tight", pad_inches=0.1)
    plt.close()

pixels = (350, 350) #FOV que precise, nesse caso a gente usou do SPARC

for name, ra, dec in galaxies:
    coord = SkyCoord(ra=ra * u.deg, dec=dec * u.deg, frame='icrs')
    saved = False

    for survey in surveys_priority:
        try:
            images = SkyView.get_images(
                position=coord,
                survey=[survey],
                coordinates="J2000",
                pixels=pixels
            )
            if images and images[0]:
                data = images[0][0].data
                save_finder_chart(data, name, ra, dec, output_dir)
                print(f" {name} – from {survey}")
                saved = True
                break
        except Exception as e:
            print(f" {name}: {survey} failed – {e}")

    if not saved:
        print(f" No image found for {name}")

print(f"\n All finder charts saved to: {output_dir}")

#Precisa de internet né
```
